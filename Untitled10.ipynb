```python
# Ensure necessary libraries are imported
from sklearn.model_selection import train_test_split

# Assuming df is a DataFrame and 'TV' and 'sales' are column names
X = df[['TV']]
Y = df[['Sales']]

# Splitting the data into training and testing datasets
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=0)

# Displaying the results
print("X_train:")
print(X_train)
print("\nY_train:")
print(Y_train)
print("\nX_test:")
print(X_test)
print("\nY_test:")
print(Y_test)
```

```python
# Import LinearRegression from sklearn
from sklearn.linear_model import LinearRegression

# Create a LinearRegression model
ML = LinearRegression()

# Fit the model using the training data
ML.fit(X_train, Y_train)
```

```python
X_train_prediction=ML.predict(X_train)
print(X_train_prediction)
```

```python
print(Y_train)
```

```python
print(ML.coef_)
print(ML.intercept_)
```

```python
X_test_prediction = ML.predict(X_test)
print(X_test_prediction)
```

```python
print(Y_test)
```

```python
import matplotlib.pyplot as plt

# Assuming you have already made predictions on the training data
X_train_prediction = ML.predict(X_train)

# Plot histogram comparison between original and predicted values
plt.figure(figsize=[7,5])

# Plot the histogram for original values
plt.hist(Y_train, bins=15, alpha=0.5, label='Original values')

# Plot the histogram for predicted values
plt.hist(X_train_prediction, bins=15, alpha=0.5, label='Predicted values')

# Adding labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram Comparison: Original vs Predicted values")
plt.legend()

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming you have already made predictions on the test data
X_test_prediction = ML.predict(X_test)

# Plot histogram comparison between original and predicted values
plt.figure(figsize=[7,5])

# Plot the histogram for original values
plt.hist(Y_test, bins=15, alpha=0.5, label='Original values')

# Plot the histogram for predicted values
plt.hist(X_test_prediction, bins=15, alpha=0.5, label='Predicted values')

# Adding labels and title
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram Comparison: Original vs Predicted values")
plt.legend()

# Show plot
plt.show()
```
